```python
import argparse
import json
from typing import List
import random
from sklearn.metrics import accuracy_score
```

```python
# Parse the input file from JSONL to a list of dictionaries.
def read_jsonl_lines(input_file: str) -> List[dict]:
    with open(input_file) as f:
        lines = f.readlines()
        return [json.loads(l.strip()) for l in lines]
    
def calc_accuracy(labels_file, output_file):
    with open(labels_file, "r") as f:
        labels = [int(line.rstrip()) for line in f]
        
    with open(output_file, "r") as o:
        preds = [int(line.rstrip()) for line in o]
        
    acc = accuracy_score(labels, preds)
    return acc
```
